##**OOP**
OOP is a style of programming that focuses on using objects to design and build applications. Think of an object as a model of the concepts, processes, or things in the real world that are meaningful to your application.   C++, Java, C# (and Python) are examples of OOP languages.

##**Objects**
In programming terms, an object is a self-contained component that contains **fields** (**properties/attributes/state**, such as variables, reference variables, and constants) and **methods** (**operations/ behaviours**) needed to make a certain type of data useful.  An object’s fields are what it knows and its methods are what it can do. 

Objects are the fundamental building blocks of applications from an object-oriented perspective. Objects of many different types are created in application using OOP. Each different type of object comes from a specific class of that type.

####**Why Bundle Into Software Objects?**

**Modularity**: The source code for an object can be written and maintained independently of the source code for other objects. Once created, an object can be easily passed around inside the system.

**Information-hiding**: By interacting only with an object's methods, the details of its internal implementation remain hidden from the outside world. 

**Code Reuse**: If an object already exists (perhaps written by another software developer), you can use that object in your program (with proper citation/permission of course!). This allows specialists to implement/test/debug complex, task-specific objects, which you can then trust to run in your own code.

**“Pluggability”** and debugging ease: If a particular object turns out to be problematic, you can simply remove it from your application and plug in a different object as its replacement. This is analogous to fixing mechanical problems in the real world. If a bolt breaks, you replace it, not the entire machine.

**Testing**: It will make testing of certain functionality easier in many caes.

##**Classes, Instances, and Instantiation**
Classes and objects are the two main aspects of object oriented programming.

\

A **class** is a blueprint, template or set of instructions and attributes to build a specific type of object. \

\
Every **object** is **instantiated** (built or created) from a class. Normally, each class should be designed and programmed to accomplish one main, and only one, thing or type of object.  Because each class should be designed to have only a single main responsibility, many classes are used to build an entire application.

\
 **A class is a data type**.



\
An **instance** is a specific object built or created from a class. You can create multiple (object) instances of a specific class. An instance is often assigned to a reference variable that is used to access all of the instance's properties/fields and methods. When you make a new instance (of an object), the process is called **instantiation**

(Note: In some languages, it is typically done using the **new** keyword).

##**Initializing Properties of an Object**
\
###**Instance Attributes**
The properties (attributes) that all objects of a class must have are defined in a method called **`.__init__()`** .  The init method set the intial state of newly created object by assigning values. The init method is a "dunder" method. 

*(Note: in many languages this is done by something called a "constructor".)*

\
\
Let's add some **instance attributes** (properties) to the Dog class.



```python
class Dog:
    def __init__(self, name, age):
        self.name = name
        self.age = age
```

While instance attributes will **vary** depending on the object instances, such as dogs having differernt names and ages, but we can also have class attributes.

###**Class Attributes**
These are attributes that have the same value for all instances of a class. For examples, a Dog class could have a species attribute with the value canine. Or a Human class could have a species attribute with a value Home sapien.

\
Class attributes are defined directly under the class name declaration and they must be initialized. They define properties that are the SAME for every instance of a class.

\
Let's add a class attribue to the Dog class:

```python
class Dog:
    # class attributes underneath class name
    species = "Canis familiaris"

    # initializing instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age
```


\
Note: When Python creates a new instance and passes it to the first parameter of `.__init__()`. This essentially removes the `self` parameter, so you only need to worry about the other parameters.

##**Instantiating an Object**
Creating a new object from a class is called **instantiating** an object. You can instantiate a new object by typing the name of the class followed by `( )`. Depending on whether or not parameters are required, you may have to pass arguments to the class' "constructor".

\
We'll create a new instance of an object of type Dog and print out its memory address. 

\

```python
class Dog:
    # class attributes underneath class name
    species = "Canis familiaris"

    # initializing instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

# end of class



# instaniate an instance of a Dog object
Dog("Buddy", 10)

# instaniate another instance of a Dog object
Dog("Doug", 7)

```

###**Instance Methods**

Instance methods are functions defined inside a class and can only be called from an instance of that class. The instance method's first parameter is always `self`.

\
Let's add some instance methods to the Dog class we have been working on:

\


```python
class Dog:
    # class attributes underneath class name
    species = "Canis familiaris"

    # initializing instance attributes
    def __init__(self, name, age):
        self.name = name
        self.age = age

    # instance methods
    def description(self):
        return (f"I'm {buddy.name} and I'm {buddy.age} years-old.")

    def speak(self, sound):
        return f"{self.name} says {sound}"

    # speical instance method, which is a "dunder" method like __init__() 
    # this method is called when the print() method is used on the instance name
    def __str__(self):
        return (f"\nI'm {buddy.name} and I'm a dog.\n")
        
# end of class




# instaniate an instance of a Dog object and assign it to a variable that will refer (or point) to the object
buddy = Dog("Buddy", 10)

print(buddy.description())
print(buddy.speak("bark"))

print(buddy)

   

```

###**Accessor and Mutator Methods -- "Getters" and "Setters"**


**Data Encapsulation (Data Hiding)**:\
We are going to get into this more detail when we cover the pillars of OOP, but the gist of it is that internal data of an object must be kept private. It needs to be hiding from the rest of the code. 

\
But there should be some methods in the class interface that can permit the user of an object to access and alter the data (that is stored internally) in a controlled manner. 

\
Note: Python does not have keywords such as private, public, and protected like some other OOP langages, but many programmers follow an accepted convention to specify what should be private or for internal use. They may put a single underscore before the variable name.    e.g. `_income`  
      
Others think it's unnecssary...


\
Most OOP languages have methods that can be called to **access or alter the state** (field/attribute values) in an object instance.

These are often called accessor and mutator methods, although names may vary depending on the language.   *E.g. "getters", "setters"*

\

####**Accessor** method: 
This method is used to access the state of the object i.e, the data hidden in the object can be accessed from this method. However, this method cannot change the state of the object, it can only access the data hidden. We often name these methods with the word **get**.

\
####**Mutator** method: 
This method is used to mutate/modify the state of an object i.e, it alters the hidden value of the data variable. It can set the value of a variable instantly to a new value. This method is also called as update method. Moreover, we often name these methods with the word **set**.

\

Let's take a look at an example:

```python
class Dog:
    # class attributes underneath class name
    species = "Canis familiaris"

    # initializing instance attributes
    def __init__(self, name, age, gender):
        self.name = name
        self.age = age
        # private attribute gender
        self._gender = gender

    # accessor method
    def get_gender(self):
        return self._gender

    # mutator method
    def set_gender(self, gen):
        self._gender = gen
# end of class


# instaniate an instance of a Dog object and assign it to a variable that will refer (or point) to the object
buddy = Dog("Buddy", 10, "male")

# use the mutator method and then accessor method
buddy.set_gender("female")
print(buddy.get_gender())
```
